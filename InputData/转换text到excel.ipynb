```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def preprocession(path,sep="/n",engine='python'):
    ds=pd.read_table(path,
             sep=sep,engine=engine,header=0)
    ds["minRange"]=0
    ds["MaxRange"]=0
    ds["probability"]=0
    ds["Mode"]=0
    ds["MidRange"]=0
    ds["Median"]=0
    ds["id1"]=0
    ds.columns=['id',
       'minRange', 'MaxRange', 'probability', 'Mode', 'MidRange', 'Median',
       'id1']
    for i in range(len(ds)):
        if ds["id"][i]=="MinRange\tMaxRange\tprobability" or ds["id"][i]=="Mode\tMidRange\tMedian":
            ds=ds.drop([i])
    ds.index=np.arange(len(ds))
    ds.dropna(how="any",inplace=True)

    for i in range(len(ds)):
        if "Identification_number" in ds["id"][i]:
            ds["id1"][i]=ds["id"][i]
            ds["minRange"][i]=ds["id"][i+1].split("\t")[0]
            ds["MaxRange"][i]=ds["id"][i+1].split("\t")[1]
            ds["probability"][i]=ds["id"][i+1].split("\t")[2]
            ds["Mode"][i]=ds["id"][i+2].split("\t")[0]
            ds["MidRange"][i]=ds["id"][i+2].split("\t")[1]
            ds["Median"][i]=ds["id"][i+2].split("\t")[2]
    for i in range(len(ds)):
        if "Identification_number" not in ds["id"][i]:
            ds=ds.drop(index=[i])
    ds=ds.drop(columns="id1")
    ds.index=np.arange(len(ds))
    for i in range(len(ds)):
        ds.iloc[i,0]=ds["id"][i].split(": ")[1]
    ds.set_index("id",inplace=True)
    return ds
```
